```python
import os
from dotenv import load_dotenv
from gridstatus import Ercot, Markets
from gridstatusio import GridStatusClient
from datetime import datetime, timedelta

# Load .env
load_dotenv()

API_KEY = os.getenv("GRIDSTATUS_API_KEY")
if API_KEY is None:
    raise RuntimeError("GRIDSTATUS_API_KEY not found in environment variables.")

client = GridStatusClient(api_key=API_KEY)
grid = Ercot()
QUERY_LIMIT = 100

def get_load_actuals(client):
    """Get ERCOT load actuals for a specific date range."""
    df = client.get_dataset(
    dataset="ercot_load",
    start="2021-1-1",
    end="2021-1-7",
    timezone="market",
    )
 
    return df

def get_net_load_actuals(client):
    """Get ERCOT load actuals for a specific date range."""
    df = client.get_dataset(
    dataset="ercot_net_load",
    start="2021-1-1",
    end="2022-12-31",
    timezone="market",
    )
 
    return df


load_actuals_df = get_load_actuals(client)
print(load_actuals_df.head())
```
