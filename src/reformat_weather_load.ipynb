```python
# Grab weather data from clean_data/cleaned_DFW_weather.csv
import pandas as pd
weather_df = pd.read_csv('../clean_data/cleaned_DFW_weather.csv')
print(weather_df.head())
```

```python
# set datetime as index
weather_df['datetime'] = pd.to_datetime(weather_df['datetime'])
weather_df.set_index('datetime', inplace=True)
print(weather_df.index)
```

```python
# Look for na values
print(weather_df.isna().sum())
```

```python
# Resample to hourly data â€” only numeric columns (avoid object/string columns)
# Use lowercase 'h' (future-proof) and numeric_only to skip non-numeric columns
weather_hourly = weather_df.resample('h').mean(numeric_only=True)
print(weather_hourly.head())
```

```python
weather_hourly.drop(columns=['STATION', "SOURCE", 'LATITUDE', 'LONGITUDE', 'ELEVATION'], inplace=True)
```

```python
weather_hourly.head()
```

```python
# save to csv
weather_hourly.to_csv('../clean_data/weather_hourly.csv')
```

```python
# grab ercot_load_actuals_2021_2024.csv
ercot_load = pd.read_csv('../clean_data/ercot_load_actuals_2021_2024.csv')
print(ercot_load.head())
```

```python
# keep only interval start local
ercot_load.drop(columns=['interval_end_local', 'interval_start_utc', 'interval_end_utc'], inplace=True)
```

```python
ercot_load['interval_start_local'] = (
    pd.to_datetime(ercot_load['interval_start_local'], errors='coerce', utc=True)
      .dt.tz_convert('America/Chicago')
      .dt.tz_localize(None)
)
ercot_load.set_index('interval_start_local', inplace=True)
ercot_hourly = ercot_load.resample('h').mean(numeric_only=True)
```

```python
ercot_hourly.head()
```

```python
# save load data
ercot_hourly.to_csv('../clean_data/hourly_load_act.csv')
```
